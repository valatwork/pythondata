```python
import matplotlib.pyplot as plt
```

```python
heights = [137,140,142,145,147,150,152,155,157,160]
f_weights = [28.5,30.8,32.6,34.9,36.4,39,40.8,43.1,44.9,47.2]
m_weights = [34.9,38.1,33.5,35.8,46.7, 42.8,43.1,45.8,50.8,58.9]
```

```python
plt.scatter(heights, f_weights)
```

```python
plt.scatter(f_weights, heights, marker="*")
plt.scatter(m_weights, heights, marker="d")
```

```python
plt.scatter(heights, f_weights,  marker="*", label="Female")
plt.scatter(heights, m_weights, marker="d", label="Male")
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.legend()
```

```python
labels = ["Turkey", "Potatoes", "Pumpkin Pie", "Stuffing"]
prices = [25.99,3.24,9.50, 6.99]
```

```python
plt.pie(prices)
```

```python
plt.pie(prices, labels=labels)
```

```python
plt.pie(prices, labels=labels, autopct="%1.3f%%")
```

```python
plt.pie(prices, labels=labels, autopct="%1.1f%%")
plt.show()
```

```python
plt.pie(prices, labels=labels, autopct="%1.1f%%", shadow=True)
plt.show()
```

```python
plt.pie(prices, labels=labels, autopct="%1.1f%%", shadow=True, startangle=90)
plt.show()
```

```python
plt.pie(prices, labels=labels, autopct="%1.1f%%", shadow=True, explode=(0,0,0.1,0))
plt.show()
```
