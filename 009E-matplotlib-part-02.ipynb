```python
# The figure should be 8x6
# The two colors used are `#74b9ff` and `#7ed6df`
import matplotlib.pyplot as plt
resorts = ["Crested Butte", "Aspen", "Telluride", "Wolf Creek"]
snow_last_24 = [29,15,15,23]
snow_next_24 = [16,6,19,30]
```

```python
plt.figure(figsize=(8,6), dpi=200)
plt.title("CO Snow Storm Report")
plt.bar(resorts, snow_last_24, color="#7ed6df", label="Last 24hrs")
plt.bar(resorts, snow_next_24, bottom=snow_last_24,  color="#74b9ff", label="Next 24hrs")
plt.legend()
plt.ylabel("Inches")
```
