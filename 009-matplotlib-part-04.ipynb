```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
nums = np.arange(5)
```

```python
plt.figure(figsize=(10,4))
plt.style.use('ggplot')
plt.plot(nums,nums)
```

```python
plt.subplot(1,3,1)
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,3,1)
plt.plot(nums,nums)
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,3,1)
plt.plot(nums,nums)
plt.subplot(1,3,2)
plt.plot(nums,nums**2, color="blue")
plt.subplot(1,3,3)
plt.plot(nums,nums**3, color="yellow")
```

```python
plt.figure(figsize=(7,4))
plt.subplot(1,3,1)
plt.plot(nums,nums)
plt.subplot(1,3,2)
plt.plot(nums,nums**2, color="blue")
plt.subplot(1,3,3)
plt.plot(nums,nums**3, color="yellow")
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,3,1)
plt.plot(nums,nums)
plt.title("X")
plt.subplot(1,3,2)
plt.plot(nums,nums**2, color="blue")
plt.title("X Squared")
plt.subplot(1,3,3)
plt.plot(nums,nums**3, color="yellow")
plt.title("X Cubed")
```
